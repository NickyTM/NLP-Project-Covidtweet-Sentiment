```python
import re
import numpy as np
import pandas as pd
from os import path
from PIL import Image
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator
import matplotlib.pyplot as plt
import preprocessor as p
```

```python
data = pd.read_csv("data/preprocess_train.csv", encoding = 'latin1')
data = data.dropna()
data['tokens'] = data['tokens'].apply(lambda x: re.sub(r'[^\w]',' ', x))
print(data.shape)
display(data.head())
```

```python
testdata = pd.read_csv("data/preprocess_test_final.csv", encoding = 'latin1') 
testdata = testdata.dropna()
testdata = testdata.drop(["sent_token"], axis=1)
testdata['tokens'] = testdata['tokens'].apply(lambda x: re.sub(r'[^\w]',' ', x))
print(testdata.shape)
display(testdata.head())
```

```python
tokentext = " ".join(tweet for tweet in data["tokens"])
print ("There are {} words in the combination of all tweet tokens.".format(len(tokentext)))
token_wordcloud = WordCloud(max_font_size=50, max_words=100, background_color="white").generate(tokentext)

plt.imshow(token_wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show()
```

```python
testtokentext = " ".join(tweet for tweet in testdata["tokens"])
print ("There are {} words in the combination of all tweet tokens.".format(len(testtokentext)))
testtoken_wordcloud = WordCloud(max_font_size=50, max_words=100, background_color="white").generate(tokentext)

plt.imshow(testtoken_wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show()
```
